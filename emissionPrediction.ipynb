LINEAR REGRESSION MODEL

```python
nox_df = df.copy()
nox_df = nox_df.drop("CO", axis = 1) # drop the target variable CO
```

```python
from sklearn.model_selection import train_test_split

X = nox_df.iloc[:, 0:-1]# -1'e kadar ,-1 hariç tüm sütunlar
y = nox_df["NOX"]
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size = 0.1)
```

```python
X
```

Model-1:Linear regression

```python
import math
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score,mean_squared_error
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression ()
linear_model.fit(X_train, y_train)
y_pred = linear_model.predict(X_test)
mae_linear = mean_absolute_error(y_test, y_pred)
mape=mean_absolute_percentage_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred,force_finite=False)
print("MAE on test set is:", round(mae_linear, 3))
print("MAPE on test set is: ",round(mape,3))#yüksek doğruluk derecesi,<%10
print("R2 on test set is: ",round(r2,3))
print("MSE on test set is: ",mse)
print("RMSE on test set is: ",math.sqrt(mse))
```

Model-2:RandomForest

```python
from sklearn.ensemble import RandomForestRegressor
# fit the model
model = RandomForestRegressor(random_state =0)
model = model.fit(X_train, y_train)
# MAE on test set again
from sklearn.metrics import mean_absolute_error
from sklearn import metrics 

y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)

print("MAE on test set:", round (mae, 3))
```

Görüldüğü gibi Random Forest Modeli bu veri kümesinde iyi çalışıyor ve ayrıca Linear Regression modelinden daha iyi.(MAE:2.557)
